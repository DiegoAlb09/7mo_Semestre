### Practica No. 9. La búsqueda local es una técnica utilizada para resolver problemas de optimización donde la solución óptima se busca moviendose iterativamente entre soluciones vecinas en el espacio de búsqueda.
### Busqueda Greedy (Busqueda Avara - Algoritmo Bean Search) El algoritmo Bean Search es una técnica simple para buscar el máximo de una función. El enfoque es ir "reduciendo" el área donde creemos que está el máximo mediante una exploración por partes. En esta practica implementaremos el algoritmo Bean Search para buscar el optimo global de una función en 2D: f(x)=-x^2+10 y la graficamos. 

#### Paso 1: Importar Librerias

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Paso 2: Definir la función objetivo

```python
def objective_function(x):
    return -x**2 + 10
```

#### Paso 3: Función para implementar el algoritmo Bean Search

```python
def bean_search(a, b, num_puntos, num_iteraciones):
  for i in range(num_iteraciones):
    #Dividir el intervalo en puntos equidistantes
    puntos = np.linspace(a, b, num_puntos)

    #Evaluar la función en los puntos
    valores_funcion = [objective_function(p) for p in puntos]

    #Encontrar el punto con el valor más alto
    indice_max = np.argmax(valores_funcion)
    x_max = puntos[indice_max]

    #Reducir el intervalo alrededor del mejor punto
    longitud_intervalo = (b - a) / 4
    a = x_max - longitud_intervalo / 2
    b = x_max + longitud_intervalo / 2

    #Mostrar el intervalo actual
    print(f'Iteración {i}: Mejor x = {x_max:.4f}, f(x) = {objective_function(x_max):.4f}')

  return x_max, objective_function(x_max)
```

#### Paso 4: Definir los parametros iniciales

```python
a_inicial = -5
b_inicial = 5
num_puntos = 100
num_iteraciones = 10
```

#### Paso 5: Ejecutar el algoritmo Bean Search

```python
x_opt, f_opt = bean_search(a_inicial, b_inicial, num_puntos, num_iteraciones)
```

#### Paso 6: Graficar la función 

```python
x_valores = np.linspace(-5, 5, 400)
y_valores = objective_function(x_valores)
plt.plot(x_valores, y_valores, label='f(x) = -x^2 + 10')
plt.scatter(x_opt, f_opt, color='red', label='Máximo encontrado')
plt.title('Bean Search')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
